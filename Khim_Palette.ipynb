```python
from google.colab import drive
import matplotlib.pyplot as plt
from PIL import Image

file_path = '/content/drive/MyDrive/Khim_imageclass/'
file_name = 'call.jpg'
img = Image.open(file_path + file_name )

plt.imshow(img)
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


img = cv2.imread(file_path + 'call.jpg').astype(np.float32) / 255.


highlight = np.zeros_like(img)

cv2.circle(highlight,
           center=(250, 300),
           radius=80,
           color=(0.678, 0.95, 1),
           thickness=-2)
cv2.circle(highlight,
           center=(100, 150),
           radius=50,
           color=(0.77, 1, 0.6784),
           thickness=-2)
cv2.circle(highlight,
           center=(410, 290),
           radius=30,
           color=(1, 0.67, 0.97),
           thickness=-2)
cv2.circle(highlight,
           center=(410, 290),
           radius=30,
           color=(1, 0.67, 0.97),
           thickness=-2)
cv2.circle(highlight,
           center=(550, 300),
           radius=100,
           color=(0.839, 0.6784, 1),
           thickness=-2)


reflect_img = cv2.addWeighted(img, 0.85, highlight, 0.55, 0)

reflect_img_rgb = cv2.cvtColor(reflect_img, cv2.COLOR_BGR2RGB)


plt.imshow(reflect_img_rgb)
plt.axis('off')


plt.savefig(file_path + 'Khim_0986655541_call.png', bbox_inches='tight', dpi=300)
plt.show()
```
