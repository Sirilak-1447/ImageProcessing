```python
from google.colab import drive
import matplotlib.pyplot as plt
from PIL import Image

file_path = '/content/drive/MyDrive/Khim_imageclass/'
file_name = 'Tongta_Gq_Khim.jpg'
img = Image.open(file_path + file_name )

plt.imshow(img)
plt.show()
```

```python
print('Loaded Image:')
print(f"Size: {img.size}")
print(f"Mode: {img.mode}")
```

Split color channels

```python
img_r, img_g, img_b = img.split()

plt.figure(figsize=(10, 10))  #(W,H) in inches
rows = 2;   cols = 2;  index = 1

plt.subplot(rows, cols, index); index += 1
plt.imshow(img_r, cmap='Reds')
plt.title('Red')

plt.subplot(rows, cols, index); index += 1
plt.imshow(img_g, cmap='Greens')
plt.title('Green')


plt.subplot(rows, cols, index); index += 1
plt.imshow(img_b, cmap='Blues')
plt.title('Blue')


# plt.subplot(rows, cols, index); index += 1
# plt.imshow(img_b, cmap='Greys')
# plt.title('Grey')

```

```python
scale = 2
img_sampled = img.resize((img.width // scale, img.height // scale))


num_color = 10
img_quantized = img.quantize(colors = num_color)

plt.figure(figsize=(10, 10))
rows = 2;   cols = 2;  index = 1

plt.subplot(rows, cols, index); index += 2;
plt.imshow(img, cmap=None)
plt.title('Original')

plt.subplot(rows, cols, index); index += 1;
plt.imshow(img, cmap=None)
plt.title('Sampled')

plt.subplot(rows, cols, index); index += 1;
plt.imshow(img_sampled, cmap=None)
plt.title('Quantized')
```
